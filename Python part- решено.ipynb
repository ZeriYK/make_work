```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib as plt
pd.set_option("display.max_columns",21)
pd.set_option("display.max_rows",5000)
```

```python
data = pd.read_csv("bets.csv")
data.head()
```

```python
data_two = pd.read_csv('events.csv')
data_two
```

```python
merg = pd.merge(data,data_two,on = 'event_id',how = 'inner')
merg = merg.set_index(["category"], inplace = False)
merg = merg.sort_values(by = 'sport',ascending = False)
merg['sport'].unique() # уникальные значения для выявления какие типы в принципе существуют 
```

```python
#merg =merg.set_index(['sport'], inplace = False)
# в этом блоке кода будет выполняться все заданные условия: 
# (1)Ставка должна быть рассчитана не позднее 12:00 15.03.2022.
# (2)Ставки вида «система» не учитываются.
# (2)Ставки CashOut, возвраты и FreeBet не участвуют в Бонусе.
# (2) сделали ставку в пре-матч 
query = merg[merg['sport'] == 'E-Sports'] # отсортируем столбец спорт по категории, которая нам интересна. В данном случае Киберспорт
# 1 
query = query[(query['accept_time'] >= '2022-03-14 12:00:00.800')&(query['accept_time'] <= '2022-03-15 12:00:00.800')]\
                                                                    .sort_values(by = 'accept_time',ascending = True)
# 2
query = query[(query['bet_type'] != "System") & (query["is_free_bet"] != True)&\
                (query['result'] != "Cashout")&\
                (query['event_stage'] != "Live")]
```

```python
#постановка коэффицента и минимальной суммы
query = query[(query['accepted_odd'] >= 1.5)&(query['amount'] >= 10)]
file = query.to_excel(excel_writer = "Z:\excel.xlsx")
```
